```python
import re
import pandas as pd
from collections import defaultdict
import spacy
spacy.cli.download("en_core_web_sm")
import logging # Setting up the loggings to monitor gensim
from gensim import corpora
from gensim.models import LdaModel
from gensim.parsing.preprocessing import preprocess_string, strip_numeric, strip_punctuation, remove_stopwords, strip_short, stem_text

logging.basicConfig(format="%(levelname)s - %(asctime)s: %(message)s", datefmt= '%H:%M:%S', level=logging.INFO)
```

```python
df = pd.read_csv("/content/Extracted_text_51 (2).csv")
```

```python
df.columns = ['PMID', 'DOI', 'Authors', 'Title','Content']

df.to_csv("your_file.csv", index=False)
```

```python

df = df[['PMID', 'DOI', 'Authors', 'Title','Content']]
```

```python
df.head(10)
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
def preprocess(text):
    result = []
    for token in preprocess_string(text):
        if not re.match(r'^[0-9]+$', token): # Ignore numeric tokens
            result.append(token)
    return result

df['Content'] = df['Content'].apply(preprocess)
```

```python
dictionary = corpora.Dictionary(df['Content'].tolist())
```

```python
corpus = [dictionary.doc2bow(doc) for doc in df['Content'].tolist()]
```

```python
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10, random_state=100, chunksize=100, passes=10)
```

```python
def print_topics(model):
    for topic in model.print_topics(-1):
        print(topic)

print_topics(lda_model)
```
